# **Handle Missing Values:**

When data is missing from a dataset, it may be represented by other values such as '?' question marks, '-' dashes or blank spaces.

These values can create problems during data analysis as they cannot be recognized as missing data by most software packages and can interfere with calculations and statistics.

By converting these values to NaN (Not a Number), we can standardize the representation of missing data and make it easier to handle and process during data analysis. NaN values are recognized by most software packages as missing data and can be easily removed, imputed or handled in other ways during data analysis. NaN stands for "Not a Number". It is a special floating-point value in Python that represents an undefined or unrepresentable value. In the context of datasets, NaN values typically represent missing or undefined values.

```python
df.isna().sum()
```

**df.isna().sum()**

 This code first marks all the empty cells in the table. Then, for each column, it counts how many empty cells there are. Finally, it gives you a list showing the number of empty cells (missing values) for each column in your table.

```python
import seaborn as sns
sns.heatmap(df.isna())
```

```python
df.dropna(subset=['reviews_per_month'], axis=0, inplace=True)
```

The **df.dropna(subset=['reviews_per_month], axis=0, inplace=True)** command is a data cleaning operation that removes rows with missing values in the "price" column from the auto_df dataframe object. This operation is useful because missing values can cause errors in machine learning algorithms, and removing rows with missing values can improve the quality of the data used for analysis.

```python
# replace missing values in reviews_per_month column with mean value
df['reviews_per_month'].fillna(df['reviews_per_month'].astype(float).mean(), inplace=True)
```

```python
df.isna().sum()
```

The code will fill the missing values in the **'reviews_per_month'** column of the auto_df dataframe object with the mean value of the non-missing values in that column. This operation is useful because missing values can cause errors in machine learning algorithms, and filling in missing values with a reasonable estimate can improve the quality of the data used for analysis.

# **Outlier Removal**

```python
Q1 = df["calculated_host_listings_count"].quantile(0.25)
Q3 = df["calculated_host_listings_count"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_filtered = df[(df["calculated_host_listings_count"] >= lower_bound) &
                 (df["calculated_host_listings_count"] <= upper_bound)]
```

# **Boxplot**

**OUTLIERS:**

What Are Outliers?

Outliers are like those odd ones out in a crowd—data points that are way different from the rest!

Why Do Outliers Matter?

They can skew your analysis and affect your results. 🤯
They might indicate errors in data collection or genuine rare events.
Ignoring them can lead to misleading insights!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("AB_NYC_2019.csv")  # Update path if needed

# Set plot size
plt.figure(figsize=(8, 5))

# Create boxplot for 'price'
sns.boxplot(x=df['price'])

# Show the plot
plt.xlabel("Price")
plt.title("Boxplot of Prices in Airbnb Dataset")
plt.show()
```

**Boxplot**

There are some unusually high prices (outliers) on the right side of the plot.

Most listings are priced lower, while a few are very expensive, creating a long tail.

These outliers could skew the average price, so it might be useful to remove or limit them (e.g., removing prices above the 99th percentile).


```python
# Dealing with outliers
df = df[(df['price'] >= df['price'].quantile(0.05)) &
        (df['price'] <= df['price'].quantile(0.95))]
```

This code deals with outliers in the price column of the df dataset.

The df dataframe is filtered to remove observations whose price is below the 5th percentile or above the 95th percentile of all price values in the dataset. This means that values that are considered as extreme or unusual are removed from the dataset.

# **Heatmap**

sns.heatmap() is a function provided by the Seaborn library, which is a popular Python data visualization library based on Matplotlib. This function is specifically designed to create heatmaps, making it easy to visualize data and analyze relationships between variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your data into df

# Convert numeric columns to appropriate types
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plot correlation heatmap for numeric columns only
plt.figure(figsize=(15, 5))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

This correlation **heatmap visualizes** the relationships between different variables in the dataset.

**Price:** Has weak correlations with all other variables. The strongest correlation is with longitude (-0.15), though still quite weak.

**Number of reviews and reviews per month:** Have a strong positive correlation (0.55), which makes sense since more frequent reviews likely lead to more total reviews.

**ID and number of reviews:** Show a moderate negative correlation (-0.32), suggesting newer listings (lower ID numbers) might get fewer reviews.

**Host ID and availability (365 days)**: Have a weak positive correlation (0.2), meaning hosts with more listings tend to have slightly higher availability.

**Other variables:** Show very weak or no significant correlations with each other.

# **Lineplot**

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert the review date column to datetime
df['last_review'] = pd.to_datetime(df['last_review'])  # Adjust the column name as necessary

# Resample data to get the number of reviews per time period (e.g., daily or monthly)
reviews_over_time = df.resample('M', on='last_review').size()  # Use 'D' for daily or 'M' for monthly

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(reviews_over_time.index, reviews_over_time.values, marker='o', linestyle='-')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

**Linplot**

From 2011 to around 2017, the number of reviews remained relatively flat, with very few reviews.

Starting around 2018, there is a noticeable increase in reviews.

The most dramatic spike occurs in 2019, where the number of reviews shoots up significantly, peaking at over 12,000 before dropping again.

# **Piechart**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

reviews_by_neighbourhood = df.groupby('neighbourhood_group')['number_of_reviews'].sum()
plt.figure(figsize=(8, 6))
plt.pie(reviews_by_neighbourhood, labels=reviews_by_neighbourhood.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Number of Reviews by Neighbourhood Group')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

- **Manhattan** closely follows with **44.6%**.

Brooklyn and Manhattan dominate in terms of reviews, while Queens, Bronx, and Staten Island contribute significantly less.
